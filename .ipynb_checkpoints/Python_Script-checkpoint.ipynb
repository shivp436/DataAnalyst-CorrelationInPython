```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

# read data
df = pd.read_csv(r'C:\Users\mpshi\Desktop\Projects\DataAnalyst-CorrelationInPython\data\movies.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# checking for blank values

for col in df.columns:
    perc_null = np.mean(df[col].isnull())
    print('{} : {}%'.format(col, perc_null))
```

```python
# remove null values

# df_new = df.dropna()

# Remove rows with any null values from the original DataFrame
df.dropna(inplace=True)
```

```python
# change budget and gross into int64

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# display all values

pd.set_option('display.max_rows', None)
```

```python
# create a correct year column

df['year_correct'] = df['released'].str.split(',').str[1].str[1:5]
```

```python
# order the data

df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
# drop duplicates based on name

df = df.drop_duplicates(subset='name', keep='first')
```

```python
# Budget - high correlation with Gross
# company - high correlation
```

```python
# scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Gross Earnings')
plt.xlabel('Budget')
plt.show()
```

```python
# regression plot - budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# looking at correlation

# select only numerical data types
# methods = pearson, kendall, spearman
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
```

```python
# plot a heatmap with correlation matrix

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# numerize the string objects like name, company

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# everyhting is changed to numerics

df_numerized.dtypes
```

```python
# correlation matrix for the new df

correlation_matrix_2 = df_numerized.corr()

# plot a heatmap with correlation matrix 2

sns.heatmap(correlation_matrix_2, annot=True)

plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# actual matrix itself

correlation_matrix_2
```

```python
# unstack the pairs

corr_pairs = correlation_matrix_2.unstack()

corr_pairs
```

```python
sorted_corr_pairs = corr_pairs.sort_values(ascending=False)

sorted_corr_pairs
```

```python
# high correlation pairs

# >= 0.9 are all of the same value to themselves and the year values, so remove them

high_corrs = sorted_corr_pairs[
    (sorted_corr_pairs > 0.5) & (sorted_corr_pairs < 0.9)
]

high_corrs
```

```python
# to only see unique relations

# Create a mask for upper triangular portion
mask = high_corrs.index.to_series().apply(lambda x: x[0] < x[1])

# Select only the upper triangular values
unique_relations = high_corrs[mask]

print(unique_relations)
```
